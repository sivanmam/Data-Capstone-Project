```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x = 'Reason', data = df, palette = 'viridis')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day Of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
df['Day Of Week'].nunique()
```

```python
dmap = {0: 'Mon', 1:'Tue', 2:'Wed', 3:'Thr', 4:'Fri', 5:'Sat', 6:'Sun'}
```

```python
df['Day Of Week'] = df['Day Of Week'].map(dmap)
```

```python
df.head()
```

```python
sns.countplot(x = 'Day Of Week', data = df, hue = 'Reason', palette = 'viridis')
plt.legend(bbox_to_anchor = (1.05,1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x = 'Month', data = df, hue = 'Reason', palette = 'viridis')
plt.legend(bbox_to_anchor = (1.05,1), loc=2, borderaxespad=0.)
```

```python
byMonth = df.groupby('Month').count()
```

```python
byMonth.head()
```

```python
byMonth['lat'].plot()
```

```python
sns.lmplot(x='Month', y='twp', data=byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
```

```python
df['Date']
```

```python
df.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
traffic = df[df['Reason']=='Traffic']
ems = df[df['Reason']=='EMS']
fire = df[df['Reason']=='Fire']
```

```python
traffic.groupby('Date').count()['lat'].plot()
plt.tight_layout()
plt.title('Traffic')
```

```python
ems.groupby('Date').count()['lat'].plot()
plt.tight_layout()
plt.title('EMS')
```

```python
fire.groupby('Date').count()['lat'].plot()
plt.tight_layout()
plt.title('Fire')
```
